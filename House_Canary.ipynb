### House Canary API query

```python
# Create Zip code list for Austin

import pandas as pd
new_coders = 'Austin_Zips.csv'
new_coders_pd = pd.read_csv(new_coders)
new_coders_pd.head()
dfList = new_coders_pd['Zipcode'].tolist()

post_data = [{'zipcode':str(v)} for v in dfList]
post_data
```

```python
# House Canary API Calls

import requests
import pprint
import getpass


url = 'https://api.housecanary.com/v2/zip/affordability_ts_historical'


df_list =[]


for i in range(len(post_data)):
        params = post_data[i]
        response = requests.get(url, params=params,auth=('954EZY40TVYVWOFDWHIG', 'IH6K17s3vOB8QqSDw7dC1HWWV50ltT7c'))
        test_addresses = response.json()
        data_list=[]
        try:
            for j in range(len(test_addresses[0]['zip/affordability_ts_historical']['result']['time_series'])):
                data_list.append(test_addresses[0]['zip/affordability_ts_historical']['result']['time_series'][j])
                
            zip_df = pd.DataFrame(data_list)
            df_list.append(zip_df)
        except:
            print('no data')        

df_list
```

```python
#len(df_list)
df_list[21]

#.tail()
```

```python
df_merged= pd.DataFrame()
df_merged['month']=""
l = 0
while l < len(df_list):
    #print(df_list[l])
    df_merged = pd.merge(df_merged,df_list[l], how = 'outer')
    l +=1
df_groupby=df_merged.groupby('month').mean().reset_index()
df_groupby
```

```python
#df_groupby.dtypes
df_groupby['month'] = pd.to_datetime(df_groupby['month'], format='%Y-%m-%d')
df_groupby.loc[400:500,]
```

```python
# Build Trend Plot

import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime


x  = np.arange(0, 515, 1)

plt.scatter(x, df_groupby["afford_pmt"], marker="o", facecolors="red", edgecolors="black", alpha=0.75)
plt.title("Forty Year Trend of Average Affordability Index\n Austin, TX : 2012-2017")
plt.xlabel("Month")
plt.ylabel("Fraction of Median Income for Payment")
plt.grid()
#plt.hlines(0, 0.33, 515, alpha=0.25)
plt.xlim(455, 515)
plt.ylim(0, 0.5)
plt.savefig("Five_year_trend.png")

plt.show()
```
