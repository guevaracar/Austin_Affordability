```python
# Reorganizing the columns using double brackets
organized_df = df[['Zip','Sqft Total','List Price','Sold/Lease$/SqFt','Sold/Lease Price','Sold/Lease Date']]
organized_df.head()
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_df = organized_df.rename(columns={"Zip":"Zip Code", "Sqft Total":"Square Footage", "Sold/Lease$/SqFt":"Sold Price Per Foot","Sold/Lease Date":"Sold Date","Sold/Lease Price":"Sold Price"})
renamed_df.head()
```

```python
renamed_df.dtypes
```

```python
renamed_df.head()
```

```python
# Converting the "Sold Price" column to floats
renamed_df["Sold Price"] = renamed_df["Sold Price"].replace('','',regex=True).astype('float')
```
